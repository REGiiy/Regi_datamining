langkah 4:Klasifikasi Untung/Rugi

````python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from google.colab import files

# Assuming you're using Google Colab, this uploads the file.
uploaded = files.upload()

# Read the file into a DataFrame
# Make sure the filename matches the uploaded file
df = pd.read_excel("Datasets Superstore.xlsx")

df['Label'] = df['Profit'].apply(lambda x: 1 if x > 0 else 0)
features = df[['Sales', 'Discount', 'Quantity']].fillna(0)
labels = df['Label']

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
````

Metode klasifikasi data lain

Berikut adalah beberapa metode klasifikasi lain selain Decision Tree dan K-Nearest Neighbors(KNN) yang dapat kamu gunakan:

✅ Metode Klasifikasi Lain
1.Naive Bayes
   - Berdasarkan probabilitas dan Teorema Bayes.
   - Cocok untuk data dengan fitur independen.
   - Cepat dan efisien untuk dataset besar.

2. Random Forest
   - Ensembel dari banyak decision tree.
   - Lebih akurat dari decision tree tunggal.
   - Mengurangi overfitting dengan voting.

3. Support Vector Machine (SVM)
   - Memisahkan kelas dengan hyperplane optimal.
   - Cocok untuk data berdimensi tinggi.
   - Tidak cocok untuk dataset sangat besar jika tidak dioptimasi.

4. Logistic Regression
   - Meskipun disebut "regresi", ini metode klasifikasi.
   - Cocok untuk prediksi kelas biner dan multinomial.
🎯 Kesimpulan Kenapa Akurasi Decision Tree, KNN, dan Metode Lain Bisa Berbeda-beda:

| Faktor | Penjelasan |
|--------|------------|
| Struktur Model | Decision Tree membagi data berdasarkan aturan logika, sedangkan KNN hanya "mengikuti" tetangga terdekat tanpa pelatihan eksplisit. Model lain seperti SVM mencari pemisah matematis terbaik. |
| Cara Belajar | KNN tidak melakukan training secara eksplisit (lazy learner), sementara Decision Tree dan model lain seperti Naive Bayes dan Logistic Regression melakukan pembelajaran eksplisit (eager learner). |
| Sensitivitas terhadap Noise & Outlier | KNN sangat sensitif terhadap outlier, Decision Tree bisa overfit, sedangkan model seperti Random Forest lebih tahan terhadap noise. |
| Skalabilitas | KNN lambat di data besar karena butuh menghitung jarak setiap prediksi. Decision Tree dan Random Forest lebih cepat setelah pelatihan selesai. |
| Distribusi Data | Beberapa model (seperti Naive Bayes) bekerja sangat baik jika asumsi distribusi data terpenuhi (misal independensi fitur), tapi buruk jika tidak. |

Apakah kamu ingin saya lanjutkan dengan preprocessing dan training model klasifikasi di file tersebut? Jika ya, tolong beri tahu target variabel klasifikasinya (misal: "Segment", "Category", dll).

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# 1. Load Data
df = pd.read_excel('Datasets Superstore.xlsx')

# 2. Drop kolom yang tidak relevan
drop_cols = ['Row ID', 'Order ID', 'Customer ID', 'Customer Name', 'Product ID', 'Product Name']
df.drop(columns=drop_cols, inplace=True)

# 3. Drop missing values (jika ada)
df.dropna(inplace=True)

# 4. Pisahkan fitur dan target
X = df.drop('Segment', axis=1)
y = df['Segment']

# 5. Encode fitur kategorikal
categorical_cols = X.select_dtypes(include='object').columns
le_feat = LabelEncoder()
for col in categorical_cols:
    X.loc[:, col] = le_feat.fit_transform(X[col])

# ---CONVERT 'Order Date' to numeric representation---
# Extract numerical features from 'Order Date'
X['Order Year'] = X['Order Date'].dt.year
X['Order Month'] = X['Order Date'].dt.month
X['Order Day'] = X['Order Date'].dt.day

# Drop the original 'Order Date' column
X = X.drop('Order Date', axis=1)
# ---END OF CONVERSION---


# 6. Encode target
le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y)

# 7. Normalisasi fitur numerik
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 8. Train-Test Split (stratify biar seimbang)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded
)

# 9. Inisialisasi model
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(random_state=42)
}

# 10. Latih dan uji model
print("Model Accuracy Comparison:\n")
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name:15}: {acc:.4f}")
````
